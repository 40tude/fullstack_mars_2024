## Dealing with matrices

```python
# Define a function called my_func that takes to arguments xand yand returns: f(x) = x² + y
def my_func (x, y):
  return x**2 + y
```

```python
# testing
my_func(3, -4)
```

```python
# Use my_func to initialize a 4x4 matrix:

my_array = np.asarray([my_func(i,j) for i in range(4) for j in range(4)]).reshape(4,4)
print(my_array)
```

```python
# Use my_func to initialize a 4x4 matrix:

my_array = np.fromfunction(my_func,(4,4),dtype=int)
print(my_array)
```

```python
# Iterate other the matrix' values and for each value, find a way of computing its remainder in the integer division by 2.
def print_reminder(n):
  print(f"When dividing {n} by 2, the remainder is: {n%2}")
```

```python
# Testing
print_reminder(15)
print_reminder(10)
```

```python
myfunc_vec = np.vectorize(print_reminder)
result = myfunc_vec(my_array)
```

```python
# Once you've found the operator that allows to compute the remainder, use it to create a mask that allows to select only even numbers in your matrix. 
# Store these values into an array called even_numbers.
print(my_array)
print()

b_even = (my_array % 2 == 0)
print(b_even)
print()

even_numbers = my_array[b_even]
print(even_numbers)
```

```python
# Reshape even_numbers into a 2x4 matrix and apply the log function to its elements.
even_numbers.reshape(2, 4)
```

```python
# apply the log function to its elements.
np.log(even_numbers)
```

# A famous theorem

```python
# Generate an array named uniform_values containing 10.000 elements that are drawn from a (continuous) uniform distribution in the interval [0, 10].

uniform_values = np.random.uniform(0, 10, 10_000)
print(uniform_values)
```

```python
import plotly.express as px

px.histogram(uniform_values)
```

```python
# Now, create a loop with 1000 iterations. 
# At each iteration:
#     Draw a new sample of 10.000 values drawn from a (continuous) uniform distribution in the interval [0, 10]
#     Compute the sample's mean (hint: numpy arrays have many useful methods to compute basic statistics)
#     Store the mean value into a list named mean_values.

# In the end, you must get a list named mean_values containing 1000 elements. The cell after will allow you to visualize the distribution of the elements in mean_values.

mean_values = []
loops = 1000
for i in range(loops):
  uniform_values = np.random.uniform(0, 10, 10_000)
  mean_values.append(uniform_values.mean())

px.histogram(mean_values, nbins=50)  
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(18, 5))
# sns.histplot(mean_values, kde=True, stat="density", kde_kws=dict(cut=3), alpha=.4, edgecolor=(1, 1, 1, .4)) 
sns.histplot(mean_values, bins=50,  alpha=.4, edgecolor=(1, 1, 1, .4))
ax.set(ylim=(0, 80)) 
ax.set_title("Normal shape ?")
_ = ax.set_xlabel("Values")
```

## Matrix operations

```python
# Compute the sum A+B of the two following matrices :

A = np.array([
  [1,2,3],
  [2,1,4],
  [3,4,1],
])

B = np.array([
  [2,3,4],
  [3,2,5],
  [4,5,2],
])

C = A + B
print(C)
```

```python
# Compute the operation 2A−B :

C = 2*A - B
print(C)
```

```python
# Compute the products of matrices AB and BA

A = np.array([
  [1,0,2],
  [2,1,0],
  [1,2,1],
])

B = np.array([
  [2,0,0],
  [1,2,0],
  [0,1,2],
])

C = A@B
print(C)
print()

C = B@A
print(C)
```

```python
# Compute the products of matrices AB and BA

A = np.array([
  [1,0],
  [2,1],
  [1,2],
])

B = np.array([
  [2,0,0],
  [1,2,0],
])

C = A@B
print(C)
print()

C = B@A
print(C)
```

## Systems of linear equations

$$
\begin{align}
2x_1 + x_2 = 2 \\
x_1 + 2x_2 + x_3 = 0 \\
x_2 + 2x_3= 1
\end{align}
$$

```python
# We'd like to write the matrix representation of this system of equations, in the form A x = c
# Create a numpy array corresponding to A and another corresponding to c

A = np.array([
  [2,1,0],
  [1,2,1],
  [0,1,2],
])

c = np.array([2,0,1])
```

```python
# Use numpy to solve the system of equations

x = np.linalg.solve(A, c)
print(x)
```

```python
# Bonus question
# Let's use the concept of matrix inverse to solve the system of linear equations !
# Prove that the following matrix B is the inverse of matrix A :

B = np.array([
  [ 3/4, -1/2, 1/4],
  [-1/2,  1  ,-1/2],
  [ 1/4, -1/2, 3/4],
])

print(A@B)
```

```python
# And then, use A-1  to find the solution to the system of equation

# B is the inverse of A so 
x = B@c
print(x)
```
